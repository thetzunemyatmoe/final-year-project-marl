```python
import time
from utils import display_plot, save_reward
from learner import train
from environment import MultiAgentGridEnv
```

```python
gammas = [0.9, 0.999]
actor_lr = 0.0001
critic_lr = 0.0001
entropy_weight = 0.01
```

```python
rewards_list = []
episodes_list = []
names = []
```

```python
for gamma in gammas:
    _, rewards, episodes = train(max_episode=5000, actor_lr=actor_lr,
                                 critic_lr=critic_lr, gamma=gamma, entropy_weight=entropy_weight, path=f'gamma/{gamma}')
    rewards_list.append(rewards)
    episodes_list.append(episodes)
    names.append(str(gamma))
    time.sleep(240)
    save_reward(f'reward/discountfactor/{gamma}/reward.txt', rewards=rewards)
    print('---------------------------------')
```

```python
display_plot(rewards_list=rewards_list, episodes_list=episodes_list, names=names,
             plot_title='Comparison of Discount Factor', filename='reward_trend_gamma', save=True)
```
