```python
import time
from utils import display_plot, save_reward
from learner import train
from environment import MultiAgentGridEnv
```

```python
gamma = 0.999
actor_lr = 0.0001
critic_lr = 0.0001
entropy_weights = [0.1, 0.05, 0.01, 0.00]
```

```python
rewards_list = []
episodes_list = []
names = []
```

```python
for entropy_weight in entropy_weights:
    _, rewards, episodes = train(max_episode=5000, actor_lr=actor_lr,
                                 critic_lr=critic_lr, gamma=gamma, entropy_weight=entropy_weight, path=f'entropyweight/{entropy_weight}')
    rewards_list.append(rewards)
    episodes_list.append(episodes)
    names.append(str(entropy_weight))
    time.sleep(240)
    save_reward(
        f'reward/entropyweight/{entropy_weight}/reward.txt', rewards=rewards)
    print('---------------------------------')
```

```python
display_plot(rewards_list=rewards_list, episodes_list=episodes_list, names=names,
             plot_title='Comparison of Entropy Regularization Coefficient', filename='reward_trend_entropy_weight', save=True)
```
