```python
import time
from utils import display_plot, save_reward
from learner import train
from environment import MultiAgentGridEnv
```

```python
configs = [
    (0.0001, 0.0001),  # Actor = Critic
    (0.0001, 0.0003),  # Actor < Critic
    (0.0003, 0.0001),  # Actor > Critic
    (0.00001, 0.0001),  # Actor < Critic
    (0.0001, 0.001),  # Actor < Critic
    (0.00003, 0.0003)  # Actor < Critic
]
```

```python
gamma = 0.999
entropy_weight = 0.01
```

```python
rewards_list = []
episodes_list = []
names = []
```

```python
for index, config in enumerate(configs):
    _, rewards, episodes = train(max_episode=5000, actor_lr=config[0],
                                 critic_lr=config[1], gamma=gamma, entropy_weight=entropy_weight, path=f'learningrate/config{index}')
    rewards_list.append(rewards)
    episodes_list.append(episodes)
    names.append(f'config{index}')
    save_reward(
        f'reward/learningrate/config{index}/reward.txt', rewards=rewards)
    time.sleep(120)
    print('---------------------------------')
```

```python
display_plot(rewards_list=rewards_list, episodes_list=episodes_list, names=names,
             plot_title='Comparison of Learning rates', filename='reward_trend_learningrate', save=True)
```
